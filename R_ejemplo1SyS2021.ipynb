# Arreglo de datos en numpy

- Señales en continuo

  $x(t) \in \mathbb{R}, \mathbb{C}$

  $t \in \mathbb{R}$

- Señales discretas

  $x[t_k] \in \mathbb{R}, \mathbb{C}$  

  $t \in \{t_0,t_1,\dots,t_{K-1}\}$; $K \in \mathbb{N}$ muestras

- Señales digitales

  $x[t_k] \in \{x_0,x_1,\dots,x_{P-1}\};$ $P\in \mathbb{N}$ estados según cantidad de bits del sistema de proceso

  $t \in \{t_0,t_1,\dots,t_{K-1}\}$; $K \in \mathbb{N}$ muestras

# Simular cero y pendiente  para acondicionamiento de señales

- Se generará una función para emular el proceso de cero y pendiente en el acondicionamiento de señales.

- Las funciones principales para generar vectores equiespaciados son [linspace](https://numpy.org/doc/stable/reference/generated/numpy.linspace.html), [range](https://www.w3schools.com/python/ref_func_range.asp) y [arange](https://numpy.org/doc/stable/reference/generated/numpy.arange.html)

````
# crear vector tiempo por rango y por número de muestras
K = 200 # número de muestras
ti = -15 # tiempo incial referencia [s]
tf = 10 # tieimpo final  [s]
# vector de tiempo equiespacioado
tv = np.linspace(start=ti,stop=tf,num=K)
print(tv)
print(tv.shape)
# tiempo muestreo se define según el rango de trabajo en segundos y la cantidad de muestras
Ts = (tf - ti)/K
print(Ts)
````

## Cálculo de pendiente y del offset (cero)

$$ y[t_k] \in [y_{min},y_{max}]$$

$$y[t_k] = m x[t_k] + c$$

$$m = \frac{y_{max}-y_{min}}{x_{max}-x_{min}}$$

$$c = y[t_k] - m x[t_k]$$

- Se crea un código tipo función bajo el ambiente def, para generar el cero y pendiente según los valores $y_{max}$ y $y_{min}$ definidos por el usuario.

````
y_min = 0 # valor min después de cero y pendiente
y_max = 5 # valor max después de cero y pendiente
#crear funcion en python
def my_ceropen(x,ymin = -5, ymax = 5): #en general se pueden definir valores por defecto
  '''
  Codigo base para simular proceso de cero y pendiente
  Se ingresa arreglo de numpy y los valores min  y max despues de cero y pendiente
  '''
  xmax = max(x) # x.max()
  xmin = min(x) # x.min()
  m = (ymax - ymin)/(xmax - xmin)
  c = ymin - m*xmin
  yv = m*x + c
  return yv
````

````
yv = my_ceropen(xv,ymin=-15,ymax=20) #se llama la función definida en la celda anterior
````

````
# Proceso de acondicionamiento cero y pendiente para una funcion coseno

# se definen valores para la señal coseno
F=60 # Frecuencia
Fs=20*F # Minimo 2 veces la frecuencia de la señal, para cumplir nyquist
Ts = 1/Fs # Numero de muestras, (Ts=1/Fs) >= 2F;
T = 1/F # Periodo de la funcion coseno, T=1/F
A = 120 # Amplitud de onda de la señal
t_i = 0 # Tiempo inicial para ubicar la señal en el plano
t_f = T # se va a graficar un solo ciclo

# datos para el eje x, antidad de muestreos ubicados en un ciclo
#Para utilizar linspace es necesario definir la cantidados de datos que se necesitan en el arreglo, asi:
#cd=(t_f-t_i)/(Ts-1)
#dt = np.linspace(t_i, t_f, cd)
dt=np.arange(t_i, t_f, Ts)

# evaluacion de la funcion de señal coseno
da = A * np.cos(2 * np.pi * dt / T)  # datos de amplitud (t)

#Grafica
font1 = {'family' : 'fantasy', 'color' : 'cyan', 'size' : 12} # personalizacion de etiqueta
font2 = {'family' : 'monospace', 'color' : 'gray', 'size' : 12} # personalizacion de titulo
plt.plot(dt, da) # Funcion de graficar
plt.title('Señal Coseno', fontdict=font2)
plt.xlabel('Tiempo [s]', fontdict=font1) # etiqueta eje x
plt.ylabel('$A\sin[\omega t_Ts]$', fontdict=font1) # etiqueta eje y
plt.show()

# cero y pendiente, para reducir la escala de la señal

da = my_ceropen(da,ymin=0,ymax=20) #se llama la función definida anteriormente

# Discretizacion de la señal, se usan los datos de 'dt' porque ya cumplen el teorema de Nyquist

plt.stem(dt,da, linefmt = '-', markerfmt='', basefmt=None ) # definicion de punts (x,y) y propiedades de las lineas
plt.title('Señal Coseno Discretizada y Reescalada', fontdict=font2)
plt.xlabel('Tiempo [s]', fontdict=font1)
plt.ylabel('$A\sin[\omega t_Ts]$', fontdict=font1)
plt.show()

# Cuantizacion con sdict y argmin

#se divide la amplitud entre la cantidad de posiciones que se pueden generar con los nbits,
#Luego la funcion cdict, evalua cada muestra de amplitud respecto a las posiciones que generan los nbits
nbits = 3 #número de bits, Entre mayor sea mejor se aproxima la cuantizacion a la curva de la señal
rmin = 0  # Limite inferior de la amplitud
rmax = 20 # Limite superior de la amplitud
#Ve es el vector de los espacios para cuantizar la amplitud en discreto, depende de la cantidad de bits
ve = np.linspace(rmin,rmax,2**nbits) # (rmax-rmin)/(2^#bits - 1(porque empieza a contar desde cero))
d = cdist(da.reshape(-1,1),ve.reshape(-1,1)) #se utiliza reshape(-1,1) para generar vectores columna como lo pide cdist  (-1,1)=Vector columna; (1,-1)=Vector fila
#d es una matriz de da x 2^nbits, ej: la fila 1 incluye la distancia euclidea entre el primer valor de da y cada elemento en ve
#se requiere identificar el elemento da[j]   más cercano a y[i] para genera señal cuantizada
ind = np.argmin(d,axis=1) #el parámetro axis = 1 indica que busca la posición a lo largo de las columnas del elemento más pequeño en cada fila
# argmin, selecciona la amplitud(posicion) de el valor de ve mas pequeño, el cual seria el mas cercano

#Codigo opcional, tomado de github
# for v in ve:
#  plt.plot(dt,v*np.ones(len(dt))) #pinta toda la linea constante en cada valor en y

dc = ve[ind] #se toman los valores del vector de estados teniendo en cuenta las posiciones indicadas por la funcion argmin



plt.plot(dt,dc ) #Grafica de la señal cuantizada, en x estan los valores que permite la velocidad de muestreo, y en el eje y los valores que permite cuantizar la cantidad de bits propuestos
plt.title('Señal Cuantizada', fontdict=font2)
plt.show()


#Función pra realizar todo el proceso anterior ademas de ajustar los limites de la señal a la que soporte el conversor analogo-digital (solo sin,cos)
Tipo=bool
seno=True
coseno=False

def Discretizar(Amplitud, Tipo, Frecuencia, ti, tf,  r_min, r_max, bits, F_s  ):

 T_s = 1/F_s
 T_ = 1/Frecuencia
 V_A = Amplitud

  #d_t = np.linspace(ti, tf, T_s)
 d_t = np.arange(ti, tf, T_s)
 if Tipo :
  d_a = V_A * np.sin(2 * np.pi * d_t / T_)
 else:
  d_a = V_A * np.cos(2 * np.pi * d_t / T_)


 font1 = {'family' : 'fantasy', 'color' : 'cyan', 'size' : 12}
 font2 = {'family' : 'monospace', 'color' : 'gray', 'size' : 12}
 plt.plot(d_t, d_a)
 plt.title('Señal', fontdict=font2)
 plt.xlabel('Tiempo [s]', fontdict=font1)
 plt.ylabel('$A\sin[\omega t_Ts]$', fontdict=font1)
 plt.show()

 d_a = my_ceropen(d_a,ymin=r_min,ymax=r_max)

 plt.stem(d_t,d_a, linefmt = '-', markerfmt='', basefmt=None )
 plt.title('Señal' and Tipo and 'Discretizada y Reescalada', fontdict=font2)
 plt.xlabel('Tiempo [s]', fontdict=font1)
 plt.ylabel('$A\sin[\omega t_Ts]$', fontdict=font1)
 plt.show()

 n_bits = bits
 r_min = r_min
 r_max = r_max
 v_e = np.linspace(r_min,r_max,2**n_bits)
 d_ = cdist(d_a.reshape(-1,1),v_e.reshape(-1,1))
 i_nd = np.argmin(d_,axis=1)
 d_c = v_e[i_nd]


 plt.plot(d_t,d_c )
 plt.title('Señal Cuantizada', fontdict=font2)
 plt.show()

Discretizar(120, seno, 60, 0, 1/60, 1, 5, 3, 100*60)
````
